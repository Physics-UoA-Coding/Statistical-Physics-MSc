Ο παρακάτω κώδικας χωρίζεται σε δύο διακριτές διαδικασίες:

1) Αριθμητική επίλυση και γραφική αναπαράσταση της εξίσωσης
$$m=\tanh\left[\beta(Jqm+H)\right]$$
2) Γραφική παράσταση της
$$f=-k_BT\left[-\frac{βJm^2q}{2}+\ln\left(2\cosh\left[β(Jmq+H)\right]\right)\right]$$

```python
#Τρέξτε αυτό αλλά μην το πειράξετε
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['mathtext.fontset']='cm'
import builtins
from IPython.display import display, HTML
def print(*args, **kwargs):
    text = " ".join(map(str, args))
    html_text = f"<span style='font-family: serif; font-size: 1.1em;'>{text}</span>"
    display(HTML(html_text))
```

### 1.1 Αριθμητική επίλυση

Θα ακολουθήσουμε την πιο πρωτόγονη μέθοδο: δοκιμάζοντας τιμές για το $m$, θα ελέγχουμε εάν το δεξιό - το αριστερό μέλος της εξίσωσης κάνει $0$.

```python
k_B=1.380649e-23     #Μονάδες SI
q=4                  #2 διαστάσεις, τετραγωνικό πλέγμα
T=100                #Kelvin
H=1e-22
J=1e-21
β=1/(k_B*T)
m=-1                 #Θυμόμαστε ότι -1<m<1
δ=1e-3               #Όσο πιο μικρό τόσο το καλύτερο... Αλλά θα κάνει περισσότερη ώρα να τρέξει!
while m<=1:
  if abs(np.tanh(β*(J*q*m+H))-m)<δ:
    print(f"Πιθανή λύση, m=", round(m,3), ". Σφάλμα ", δ)
  m+=δ
```

### 1.2 Γραφική παράσταση

Σιγουρευτείτε ότι έχετε τρέξει πρώτα το προηγούμενο κελί, ώστε να έχουν αποδοθεί οι τιμές στις παραμέτρους.

```python
m=np.linspace(-1,1,1000)
T=np.tanh(β*(J*q*m+H))
plt.figure(figsize=(8,4.5))
plt.plot(m,m, '--', color='darkgray', label='$m$')
plt.plot(m,T, color='black', label='$tanh[β(Jqm+H)]$')
plt.title('Λύσεις τα σημεία τομής', fontfamily='serif', fontsize='14')
plt.xlabel('$m$', fontfamily='serif', fontsize='12')
plt.legend()
plt.show()
```

### 2.1 Γραφική παράσταση της "ειδικής" ελεύθερης ενέργειας $f$.

```python
m=np.linspace(-1,1,1000)
f=-k_B*T*(-β*J*q*m**2/2+np.log(np.cosh(β*(J*q*m+H))))
plt.figure(figsize=(8,4.5))
plt.plot(m,f, 'black', label='$f$')
plt.title(f"$f(m), \; H=${H}", fontfamily='serif', fontsize='14')
plt.xlabel('$m$', fontsize='12')
plt.ylabel('$f$', fontsize='12')
plt.show()
```
